```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cortisone_cushings = pd.read_csv('cortisone.csv')
```

```python
cortisone_healthy = pd.read_csv('cortisone-healthy.csv')
```

```python
box = plt.boxplot([cortisone_cushings , cortisone_healthy], labels = ['Cushings','Healthy'], patch_artist = True, positions=[1,1.7])
plt.title('Cortisone Median Values |Cushings vs Healthy|')

colors = ['#E3242B','#ADC865']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```
